```python
# Task 1: Exploring and Visualizing a Simple Dataset (Iris)

# Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load Dataset
df = sns.load_dataset("iris")  # alternatively use pd.read_csv("iris.csv")
print("Dataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nFirst 5 Rows:\n", df.head())

# Step 3: Summary Statistics
print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:\n", df.describe())

# Step 4: Data Visualization

# 4.1 Scatter Plot (Sepal vs Petal Length by Species)
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()

# 4.2 Histograms
df.hist(figsize=(10, 6), color='skyblue')
plt.suptitle("Feature Distributions")
plt.tight_layout()
plt.show()

# 4.3 Box Plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Features")
plt.show()
```
